```python
import pandas as pd
import json
import numpy as np
```

```python
# read data
df = pd.read_csv("./liber777_OZ/liber_777_csv_utf8.csv", sep=",")
df = df.transpose()
```

```python
# inspect
df.head()
```

```python
# This handles both None and NaN values
df = df.replace([None, pd.NA, pd.NaT, np.nan], '...')
```

```python
df = df.reset_index() # reset index
df = df.iloc[1:] # remove first row
```

```python
# dump to json string
json_string = df.to_json(orient='records', force_ascii=False)
o = json.loads(json_string)
```

```python
json_data = json.dumps(o, ensure_ascii=False).encode('utf-8').decode('unicode_escape')

# Specify the JavaScript file path
js_file_path = '../src/liber_777.js'

import json
with open(js_file_path, 'w', encoding='utf8') as f:
	f.write('export const Liber777 = ')
	json.dump(o, f, ensure_ascii=False)
	f.write(';')
```
